```python
import pandas as pd
import requests
from collections import Counter
from get_data import DATA_URLS

pd.set_option('display.max_rows', 100)
```

## Get data

```python
film_locations_url = DATA_URLS['film_locations']
park_scores_url = DATA_URLS['park_scores']
private_open_spaces_url = DATA_URLS['private_open_spaces']
properties_url = DATA_URLS['properties']

film_locations = requests.get(film_locations_url).json()
park_scores = requests.get(park_scores_url).json()
private_open_spaces = requests.get(private_open_spaces_url).json()
properties = requests.get(properties_url).json()
```

## Check contents of datasets

```python
def check_key_counts(json_data):
    print(f'number of records: {len(json_data)}')
    keys = Counter()
    for record in json_data:
        for k in record:
            keys[k] += 1
    return keys
```

```python
check_key_counts(film_locations)
```

```python
check_key_counts(park_scores)
```

```python
check_key_counts(private_open_spaces)
```

```python
check_key_counts(properties)
```
